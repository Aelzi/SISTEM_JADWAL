## Jadwal **terbagi rata**, tapi masi terdapat asisten yang **tidak dapat jadwal**

```python
import csv
import random
import pandas as pd
from collections import defaultdict

# Membaca data dari file CSV
murid_jadwal = defaultdict(list)
file_path = 'jadwal_kosong.csv'

# Membaca file CSV dengan pandas
df = pd.read_csv(file_path)

# Membaca data dari DataFrame
for index, row in df.iterrows():
    nama, hari = row['nama'], row['hari']
    murid_jadwal[nama].append(hari)

# Inisialisasi jadwal piket
hari_piket = list(set([hari for jadwal in murid_jadwal.values() for hari in jadwal]))
jadwal_piket = {hari: [] for hari in hari_piket}

# Fungsi untuk membagi jadwal piket
def bagi_jadwal_piket(murid_jadwal, jadwal_piket):
    # Menghitung berapa kali setiap murid mendapatkan jadwal piket
    murid_count = {murid: 0 for murid in murid_jadwal}

    # Menghitung berapa kali setiap hari sudah mendapatkan siswa piket
    hari_count = {hari: 0 for hari in hari_piket}

    for hari in hari_piket:
        murid_acak = list(murid_jadwal.keys())
        random.shuffle(murid_acak)

        for murid in murid_acak:
            jadwal_kosong = murid_jadwal[murid]
            if murid_count[murid] < 2 and hari in jadwal_kosong and hari_count[hari] < len(murid_jadwal) / len(hari_piket):
                jadwal_piket[hari].append(murid)
                murid_count[murid] += 1
                hari_count[hari] += 1
            if all(murid_count[murid] == 2 for murid in murid_count) and all(hari_count[hari] == len(murid_jadwal) / len(hari_piket) for hari in hari_count):
                break

        # Memastikan setiap murid mendapatkan jadwal piket
        for murid in murid_jadwal:
            if murid not in murid_count:
                for hari in hari_piket:
                    if hari in murid_jadwal[murid] and hari_count[hari] < len(murid_jadwal) / len(hari_piket):
                        jadwal_piket[hari].append(murid)
                        murid_count[murid] += 1
                        hari_count[hari] += 1
                        break

# Memanggil fungsi untuk membagi jadwal piket
bagi_jadwal_piket(murid_jadwal, jadwal_piket)

# Menampilkan jadwal piket
for hari, murid_piket in jadwal_piket.items():
    print(f'{hari}: {", ".join(murid_piket)}')

# Mengkonversi jadwal piket menjadi DataFrame
df_jadwal_piket = pd.DataFrame(jadwal_piket)

# Menyimpan DataFrame ke dalam file CSV
df_jadwal_piket.to_csv('jadwal_piket1.csv', index=False)

# Mencari nama siswa yang tidak mendapatkan jadwal piket
nama_murid = set(murid_jadwal.keys())
nama_murid_terjadwal = set([murid for jadwal in jadwal_piket.values() for murid in jadwal])
nama_murid_tidak_terjadwal = nama_murid - nama_murid_terjadwal

# Menampilkan nama siswa yang tidak mendapatkan jadwal piket
if len(nama_murid_tidak_terjadwal) > 0:
    print(f'Siswa tidak mendapatkan jadwal piket: {", ".join(nama_murid_tidak_terjadwal)}')
else:
    print('Semua siswa mendapatkan jadwal piket')

```

## SEMUA ASISTEN DAPAT JADWAL **TAPI TIDAK RATA**

```python
import csv
import random
import pandas as pd
from collections import defaultdict

# Membaca data dari file CSV
murid_jadwal = defaultdict(list)
file_path = 'jadwal_kosong.csv'

# Membaca file CSV dengan pandas
df = pd.read_csv(file_path)

# Membaca data dari DataFrame
for index, row in df.iterrows():
    nama, hari = row['nama'], row['hari']
    murid_jadwal[nama].append(hari)

# Inisialisasi jadwal piket
hari_piket = ['Senin', 'Selasa', 'Rabu', 'Kamis', 'Jumat', 'Sabtu']
jadwal_piket = {hari: [] for hari in hari_piket}

# Fungsi untuk membagi jadwal piket
def bagi_jadwal_piket(murid_jadwal, jadwal_piket):
    # Menghitung berapa kali setiap murid mendapatkan jadwal piket
    murid_count = {murid: 0 for murid in murid_jadwal}
    print(murid_count)

    # Menghitung berapa kali setiap hari sudah mendapatkan siswa piket
    hari_count = {hari: 0 for hari in hari_piket}
    print(hari_count)

    # Jumlah siswa yang mendapatkan jadwal piket dalam seminggu
    jumlah_siswa_terjadwal = len(murid_jadwal) * 2  # 2 kali dalam seminggu

    for hari in hari_piket:
        murid_acak = list(murid_jadwal.keys())
        random.shuffle(murid_acak)

        for murid in murid_acak:
            jadwal_kosong = murid_jadwal[murid]
            if murid_count[murid] < 2 and hari in jadwal_kosong and hari_count[hari] < jumlah_siswa_terjadwal:
                jadwal_piket[hari].append(murid)
                murid_count[murid] += 1
                hari_count[hari] += 1
            if all(murid_count[murid] == 2 for murid in murid_count) and all(hari_count[hari] == jumlah_siswa_terjadwal for hari in hari_count):
                break

    # Memastikan setiap murid mendapatkan jadwal piket
    for murid in murid_jadwal:
        if murid not in murid_count:
            for hari in hari_piket:
                if hari in murid_jadwal[murid] and hari_count[hari] < jumlah_siswa_terjadwal:
                    jadwal_piket[hari].append(murid)
                    murid_count[murid] += 1
                    hari_count[hari] += 1
                    break

# Memanggil fungsi untuk membagi jadwal piket
bagi_jadwal_piket(murid_jadwal, jadwal_piket)

# Menampilkan jadwal piket
for hari, murid_piket in jadwal_piket.items():
    print(f'{hari}: {", ".join(murid_piket)}')

# Mengkonversi jadwal piket menjadi DataFrame
df_jadwal_piket = pd.DataFrame.from_dict(jadwal_piket, orient='index').T

# Menyimpan DataFrame ke dalam file CSV
df_jadwal_piket.to_csv('jadwal_piket2.csv', index=False)

# Mencari nama siswa yang tidak mendapatkan jadwal piket
nama_murid = set(murid_jadwal.keys())
nama_murid_terjadwal = set([murid for jadwal in jadwal_piket.values() for murid in jadwal])
nama_murid_tidak_terjadwal = nama_murid - nama_murid_terjadwal

# Menampilkan nama siswa yang tidak mendapatkan jadwal piket
if len(nama_murid_tidak_terjadwal) > 0:
    print(f'Siswa tidak mendapatkan jadwal piket: {", ".join(nama_murid_tidak_terjadwal)}')
else:
    print('Semua siswa mendapatkan jadwal piket')
```

### TERBAGI RATA TAPI MASI TERDAPAT SISWA YANG TIDAK DAPAT SHIFT

```python
import csv
import random
import pandas as pd
from collections import defaultdict

# Membaca data dari file CSV
murid_jadwal = defaultdict(list)
file_path = 'jadwal_kosong.csv'

# Membaca file CSV dengan pandas
df = pd.read_csv(file_path)

# Membaca data dari DataFrame
for index, row in df.iterrows():
    nama, hari = row['nama'], row['hari']
    murid_jadwal[nama].append(hari)

# Inisialisasi jadwal piket
hari_piket = list(set([hari for jadwal in murid_jadwal.values() for hari in jadwal]))
jadwal_piket = {hari: [] for hari in hari_piket}

# Menyusun daftar siswa yang akan dijadwalkan berdasarkan urutan acak
murid_acak = list(murid_jadwal.keys())
random.shuffle(murid_acak)

# Membagi jadwal piket dengan lebih merata
for hari in hari_piket:
    siswa_piket = []
    
    for murid in murid_acak:
        if hari in murid_jadwal[murid] and len(siswa_piket) < len(murid_jadwal) / 2:
            siswa_piket.append(murid)
    
    jadwal_piket[hari] = siswa_piket
    
# Mengkonversi jadwal piket menjadi DataFrame
df_jadwal_piket = pd.DataFrame.from_dict(jadwal_piket, orient='index').T

# Menyimpan DataFrame ke dalam file CSV
df_jadwal_piket.to_csv('jadwal_piket4.csv', index=False)

# Menampilkan jadwal piket
for hari, murid_piket in jadwal_piket.items():
    print(f'{hari}: {", ".join(murid_piket)}')

# Mencari nama siswa yang tidak mendapatkan jadwal piket
nama_murid = set(murid_jadwal.keys())
nama_murid_terjadwal = set([murid for jadwal in jadwal_piket.values() for murid in jadwal])
nama_murid_tidak_terjadwal = nama_murid - nama_murid_terjadwal

# Menampilkan nama siswa yang tidak mendapatkan jadwal piket
if len(nama_murid_tidak_terjadwal) > 0:
    print(f'Siswa tidak mendapatkan jadwal piket: {", ".join(nama_murid_tidak_terjadwal)}')
else:
    print('Semua siswa mendapatkan jadwal piket')
```
